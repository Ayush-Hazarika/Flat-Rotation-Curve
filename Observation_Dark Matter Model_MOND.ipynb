## Dark Matter Halo Model

```python
vc, uo, u1, u2, RD, I0, I1, k0, k1, rho0, r, y, atan = sp.symbols('v_c u_o u_1 u_2 R_D I_{0}(y) I_{1}(y) k_{0}(y) k_{1}(y) rho_0 r y arctan(r/u_2)')
```

```python
a= vc**2 - uo*RD*(y)**2 *(I0*k0-I1*k1) - u1*(u2**2-u2**3/r *atan)
```

```python
## The equation for dark matter halo model
```

```python
def calculate_v(r, R_D, u_0, u_1, u_2):#NGC 3198
    y = r / (2 * R_D)
    term1 = u_0 * R_D * y**2 * (iv(0, y) * kv(0, y) - iv(1, y) * kv(1, y))
    term2 = u_1 * (u_2**2 - (u_2**3 / r) * np.arctan(r / u_2))
    return np.sqrt(term1 + term2)

r_values = np.linspace(0.1, 30, 50)
R_D = 2.63  
u_0 = 35680 
u_1 = 619  
u_2 = 6.414 

v_values = calculate_v(r_values, R_D, u_0, u_1, u_2)

# Plotting the graph
plt.plot(r_values, v_values, '-',color = 'Blue')
plt.errorbar(NGC_3198_R, NGC_3198_V, yerr=error_3198, fmt='.',color = 'Red',capsize=2)
arrow_props = dict(arrowstyle='->', linewidth=2)
plt.annotate('From the Dark matter halo model', xy=(1, 80), xytext=(5, 80), arrowprops=arrow_props,fontfamily='Times New Roman')
plt.xlabel('R(kpc)',fontfamily='Times New Roman')
plt.ylabel(r'$V_c(km/s)$', fontfamily='Times New Roman')
plt.title(r'Disk plus Dark matter Halo Model for NGC 3198: $V_c$ v/s R',fontfamily='Times New Roman')
plt.grid(True)
#plt.savefig('Disk3198.png')
plt.show()
```
